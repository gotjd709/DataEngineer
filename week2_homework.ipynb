```python
%%sql
SELECT usc.userid, st.ts, usc.channel, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts ASC) AS seq1, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts DESC) AS seq2
FROM raw_data.user_session_channel AS usc
JOIN raw_data.session_timestamp AS st 
    ON usc.sessionid = st.sessionid
LIMIT 100;
```

```python
%%sql

SELECT t1.userid, t1.channel as first_channel, t2.channel as last_channel, t1.ts, t2.ts, t1.seq1, t2.seq2
FROM (
    SELECT usc.userid, st.ts, usc.channel, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts ASC) AS seq1, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts DESC) AS seq2
    FROM raw_data.user_session_channel AS usc
    JOIN raw_data.session_timestamp AS st 
        ON usc.sessionid = st.sessionid
) as t1
JOIN (
    SELECT usc.userid, st.ts, usc.channel, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts ASC) AS seq1, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts DESC) AS seq2
    FROM raw_data.user_session_channel AS usc
    JOIN raw_data.session_timestamp AS st 
        ON usc.sessionid = st.sessionid
) as t2
    ON t1.userid = t2.userid
LIMIT 100;
```

```python
%%sql

SELECT t1.userid, t1.channel as first_channel, t2.channel as last_channel
FROM (
    SELECT usc.userid, st.ts, usc.channel, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts ASC) AS seq1, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts DESC) AS seq2
    FROM raw_data.user_session_channel AS usc
    JOIN raw_data.session_timestamp AS st 
        ON usc.sessionid = st.sessionid
) as t1
JOIN (
    SELECT usc.userid, st.ts, usc.channel, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts ASC) AS seq1, ROW_NUMBER() OVER(PARTITION BY usc.userid ORDER BY st.ts DESC) AS seq2
    FROM raw_data.user_session_channel AS usc
    JOIN raw_data.session_timestamp AS st 
        ON usc.sessionid = st.sessionid
) as t2
    ON t1.userid = t2.userid
WHERE t1.seq1 = 1 and t2.seq2 = 1
```

```python
%%sql

SELECT usc.userid, SUM(st.amount) as gross_revenue, SUM(CASE WHEN st.refunded is not TRUE then st.amount END) as net_revenue
FROM raw_data.session_transaction as st
LEFT JOIN raw_data.user_session_channel as usc
    ON usc.sessionid = st.sessionid
GROUP BY usc.userid
ORDER BY gross_revenue DESC;
```

### Assignment 4

```python
%%sql

SELECT LEFT(t.ts,7) as "month",
    usc.channel,
    COUNT(DISTINCT usc.userid) as uniqueUsers
FROM raw_data.user_session_channel as usc
LEFT JOIN raw_data.session_timestamp t 
    ON t.sessionid = usc.sessionid
LEFT JOIN raw_data.session_transaction st 
    ON t.sessionid = usc.sessionid
GROUP BY 1, 2
ORDER BY 1, 2;
```

```python
%%sql

SELECT LEFT(ts,7) "month",
    usc.channel,
    COUNT(DISTINCT userid) uniqueUsers,
    COUNT(DISTINCT (CASE WHEN amount is not NULL THEN userid END)) paidUsers,
    ROUND(paidUsers::float*100/NULLIF(uniqueUsers, 0),2) conversionRate,
    SUM(amount) grossRevenue,
    SUM(CASE WHEN refunded is not True THEN amount END) net_revenue
FROM raw_data.user_session_channel usc
LEFT JOIN raw_data.session_timestamp t ON t.sessionid = usc.sessionid
LEFT JOIN raw_data.session_transaction st ON st.sessionid = usc.sessionid
GROUP BY 1, 2
ORDER BY 1, 2;
```
